## Gradient Descent - Single variable without bias

```python
import numpy as np
import matplotlib.pyplot  as plt
import matplotlib.animation as animation
from IPython.display import clear_output
import time
```

```python
#input
X = np.arange(5)
X
```

```python
#output
Y = np.arange(0,9,2)
Y
```

```python
plt.scatter(X,Y)
plt.show
```

```python
theta = 0.0
H = theta*X
plt.scatter(X,Y)
plt.scatter(X,H)
plt.show()
```

```python
cost = np.square(np.sum(H-Y))/(2*5)
print("with theta = "+str(theta)+" cost is = "+str(cost))
```

```python
alpha = 0.02
theta = 0.0
cost_record_per_iteration = []
plt.figure(figsize=(16,9))
for i in range(0,31):
    global theta,cost_record_per_iteration
    H = theta*X
    cost = np.square(np.sum(H-Y))/(2*5)
    cost_record_per_iteration.extend([cost])
    cost_d = np.sum((H - Y)*X)/5
    if(i%5==0):
        plt.scatter(X,Y)
        plt.scatter(X,H)
        plt.xlabel("X")
        plt.ylabel("Y")
        plt.title("Cost with theta = %.6f is %.6f" %(theta, cost))
        if(i/5!=6):
            plt.subplot(2,3,i/5+1)
    theta = theta - alpha*cost_d
plt.show()
    
```

```python
#cost function vs Iteration
iteration = range(0,31)
plt.plot(iteration,cost_record_per_iteration)
plt.xlabel("Number of iteration")
plt.ylabel("cost function")
plt.show()
```
